# Project 1: Depression in Students at Universities

The study conducted aims to explore whether high academic pressure contributes to suicidal thoughts among students. The question I'd like to address is do students with higher academic pressure experience more suicidal thoughts compared to those with lower academic pressure? Given the increasing mental health concerns among students, this project seeks to analyze patterns and correlations between academic stress and mental well-being.

We will be using the Depression Student Dataset to investigate this relationship. (Dataset is from Kaggle https://www.kaggle.com/datasets/ikynahidwin/depression-student-dataset)

Dataset includes 502 student responses. Features included are the student's gender, their academic pressure and study satisfaction (ranging low 1 - high 5), their lifestyle habits (sleep and dietary habits), and other factors listed below.

# Visualization

```python
df_cleaned = dsd_df.dropna(subset=['Academic Pressure', 'Have you ever had suicidal thoughts ?'])
df_cleaned.columns = df_cleaned.columns.str.strip()
df_cleaned = df_cleaned.rename(columns={'Have you ever had suicidal thoughts ?': 'Suicidal Thoughts'})
df_cleaned['Academic Pressure'] = pd.to_numeric(df_cleaned['Academic Pressure'], errors='coerce')
df_cleaned['Academic Pressure'] = pd.Categorical(df_cleaned['Academic Pressure'], categories=[1, 2, 3, 4, 5], ordered=True)
df_filtered = df_cleaned[df_cleaned['Suicidal Thoughts'].str.lower() == 'yes']
academic_pressure_counts = df_filtered['Academic Pressure'].value_counts().reindex([1, 2, 3, 4, 5], fill_value=0)
plt.figure(figsize=(10, 6))
academic_pressure_counts.plot(kind='bar', color='royalblue')
plt.title("Suicidal Thoughts Count by Academic Pressure Level")
plt.xlabel("Academic Pressure Level (1-5)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

# Summary

Moderate academic pressure (level 3) might create the most psychological distress, possibly because students at this stage feel the strain of balancing workload expectations. We see here that students with moderate academic pressure tend to have more suicidal thoughts than students that experience both low and high academic pressure. 
Students with lower pressure (1-2) does not eliminate suicidal thoughts completely. However, they are there and we can assume that there are other factors contributing to their thoughts.
Students with higher pressure (4-5) do not show a drastic increase after moderate pressure. However, we can assume there's a possibility that they may have ways to deter those thoughts.
Overall, students with moderate to high academic pressure have more suicidal thoughts than students with lower academic pressure. However, there are possibilities that students have other factors outside of academic pressure that play a critical part in their thoughts.
